# **Difference between .loc and .iloc**

# .loc:
 - Label-based indexing.
 - Selects rows and columns based on their labels (index and column names).
 - Inclusive of both the start and end indices.
 - Can also be used with boolean indexing to select rows based on a condition.

# Example:
   df.loc[0:2, ['col1', 'col2']] #selects rows with index labels 0, 1, and 2, and columns named 'col1' and 'col2'
   df.loc[df['col1'] > 10] #select rows where the 'col1' value is greater than 10.

# .iloc:
 - Integer-based indexing.
 - Selects rows and columns based on their integer positions (starting from 0).
 - Exclusive of the end index.


# Example:
   df.iloc[0:2, 0:2]  # selects the first two rows and the first two columns (index 0 and 1 for both).
   df.iloc[[0, 2, 4], [0, 2]] # selects rows with indices 0, 2, 4 and columns at indices 0 and 2

#  Difference between drop axis=1 and drop axis =0

 The difference lies in the direction along which the operation is performed.

# loc.drop('wind_speed', axis=1)
 Drops the column labeled 'wind_speed'.  axis=1 signifies column operation.

# loc.drop(loc.index[0], axis=0)
 Drops the row at index 0. axis=0 signifies row operation.  loc.index[0] gets the label of the first row.
 In this case, it's equivalent to:
 loc.drop(0, axis=0) or even more simply: loc.drop(0) since axis=0 is the default for drop().

# In summary:
 axis=0:  Applies the operation along rows (i.e., dropping rows).
 axis=1: Applies the operation along columns (i.e., dropping columns).
