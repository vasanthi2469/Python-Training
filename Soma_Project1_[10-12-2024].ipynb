```python
# prompt: Check for missing values and duplicates in the dataset & handle the missing values

# Check for missing values
print(Retail.isnull().sum())

# Check for duplicates
print(Retail.duplicated().sum())

# Handle missing values (example: fill with mean for numerical, mode for categorical)
for col in Retail.columns:
    if pd.api.types.is_numeric_dtype(Retail[col]):
        Retail[col] = Retail[col].fillna(Retail[col].mean())
    else:
        Retail[col] = Retail[col].fillna(Retail[col].mode()[0])

# Verify if missing values are handled
print(Retail.isnull().sum())
```

```python
# prompt: Find the region contributing the most to sales.

# Assuming 'Retail' DataFrame is already loaded and cleaned

# Calculate total sales for each region
region_sales = Retail.groupby('Region')['Total Sales'].sum()

# Find the region with the highest total sales
highest_sales_region = region_sales.idxmax()

# Print the result
print(f"The region contributing the most to sales is: {highest_sales_region}")
```

# **Assignment-4: -** Timestamp Transformation
  ### **Task-2: -**  
Create new columns:

o Day of the week.

o Day of the month.

o Month.

o Year.

o Season (Winter, Spring, Summer, Fall).

```python


# Assuming 'Retail' DataFrame is already loaded and 'Order Product Date' is datetime

# Extract date components
Retail['Day of the week'] = Retail['Order Product Date'].dt.dayofweek  # Monday=0, Sunday=6
Retail['Day of the month'] = Retail['Order Product Date'].dt.day
Retail['Month'] = Retail['Order Product Date'].dt.month
Retail['Year'] = Retail['Order Product Date'].dt.year

# Define a function to determine the season based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create the 'Season' column
Retail['Season'] = Retail['Month'].apply(get_season)

Retail
```

```python
# prompt: Compute the correlation matrix for numerical columns.

import pandas as pd

# Assuming 'Retail' DataFrame is already loaded and processed

# Select only numerical columns
numerical_cols = Retail.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_cols.corr()

# Display the correlation matrix
correlation_matrix
```

# **SUMMARY: -**
This code calculates the correlation matrix for numerical columns in a Pandas DataFrame called `Retail`.

   - `numerical_cols = Retail.select_dtypes(include=np.number)`: This line filters the DataFrame to keep only columns that contain numerical data.  `select_dtypes()` is a pandas function to select columns based on their data type; `np.number` specifies that you want only numerical columns. The result is stored in the variable `numerical_cols`.

2. **Calculate Correlation Matrix:**
   - `correlation_matrix = numerical_cols.corr()`: This is the key step.  The `.corr()` method computes the pairwise correlation of columns in the `numerical_cols` DataFrame.  The correlation measures the linear relationship between two variables. A correlation of 1 means a perfect positive linear relationship, -1 means a perfect negative linear relationship, and 0 means no linear relationship. The result, a matrix of correlation coefficients, is stored in the `correlation_matrix` variable.

3. **Display Correlation Matrix:**
   - `correlation_matrix`: This simply displays the calculated correlation matrix.  The matrix shows the correlation coefficient for every pair of numerical columns in the DataFrame.

**In simpler terms:** The code takes a table of data, finds the columns with numbers, and then calculates how strongly each number column is related to every other number column. The result is a table where each cell shows the relationship between two columns.  A value of 1 means they move together perfectly; a value of -1 means they move in opposite directions perfectly; a value near 0 means they don't move together in any predictable way.

Drop a column (e.g., Region) using drop(axis=1)

# **SUMMARY: -**
This code removes the 'Region' column from the `Retail` DataFrame.

Here's how it works:

* **`Retail = Retail.drop('Region', axis=1)`**: This is the core line.
    * `Retail.drop(...)`: The `.drop()` method is used to remove rows or columns from a DataFrame.
    * `'Region'`: This is the label of the column you want to remove.  So, the code tells pandas to remove the column named 'Region'.
    * `axis=1`:  This parameter specifies that you're dropping a *column*.  `axis=0` would drop a row.  Columns are considered the "1" axis, and rows are the "0" axis in pandas.
    * `Retail = ...`: The result of the `.drop()` operation (the modified DataFrame without the 'Region' column) is assigned back to the `Retail` variable. This is important:  it *overwrites* the original `Retail` DataFrame with the modified version. If you didn't do this, the original `Retail` DataFrame would remain unchanged.
* **`Retail`**: This line simply displays the updated DataFrame, allowing you to see the changes made by the drop operation. You'll now see that the 'Region' column is no longer present in the output.

In short:  It deletes a column from a table of data. The `axis=1` is crucial to tell it to delete a column and not a row.  The reassignment (`Retail = ...`) makes the change permanent to the dataframe.
