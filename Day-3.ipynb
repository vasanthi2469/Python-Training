```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/content/credit-risk/credit.csv')

# Basic Data Exploration
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

# Calculate Statistical Measures
mean_value = df['column_name'].mean()
median_value = df['column_name'].median()
mode_value = df['column_name'].mode()
range_value = df['column_name'].max() - df['column_name'].min()
variance_value = df['column_name'].var()
std_dev_value = df['column_name'].std()
skewness_value = df['column_name'].skew()

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Range:", range_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_dev_value)
print("Skewness:", skewness_value)

# Visualize the Data
plt.hist(df['column_name'], bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Column Name')
plt.show()

plt.boxplot(df['column_name'])
plt.title('Box Plot of Column Name')
plt.show()
```

```python
import statistics
from scipy.stats import skew

# Nominal Data (e.g., categories, no intrinsic order)
nominal_data = {"Red", "Blue", "Green"}
print("Nominal Data:", nominal_data)

# Ordinal Data (e.g., categories with order)
ordinal_data = {"Low", "Medium", "High"}
print("Ordinal Data:", ordinal_data)

# Discrete Data (e.g., countable, distinct values)
discrete_data = [1, 2, 3, 4, 5]
print("Discrete Data:", discrete_data)

# Continuous Data (e.g., real numbers, measurable values)
continuous_data = [1.5, 2.3, 3.8, 4.1, 5.0]
print("Continuous Data:", continuous_data)

# Mean
mean_discrete = statistics.mean(discrete_data)
mean_continuous = statistics.mean(continuous_data)
print("Mean (Discrete):", mean_discrete)
print("Mean (Continuous):", mean_continuous)

# Mode
mode_discrete = statistics.mode(discrete_data)
print("Mode (Discrete):", mode_discrete)

# Median
median_discrete = statistics.median(discrete_data)
median_continuous = statistics.median(continuous_data)
print("Median (Discrete):", median_discrete)
print("Median (Continuous):", median_continuous)

# Range
range_discrete = max(discrete_data) - min(discrete_data)
range_continuous = max(continuous_data) - min(continuous_data)
print("Range (Discrete):", range_discrete)
print("Range (Continuous):", range_continuous)

# Skewness
skewness_discrete = skew(discrete_data)
skewness_continuous = skew(continuous_data)
print("Skewness (Discrete):", skewness_discrete)
print("Skewness (Continuous):", skewness_continuous)
```

```python
import statistics
from scipy.stats import skew
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
nominal_data = {"Red", "Blue", "Green"}
ordinal_data = ["Low", "Medium", "High"]
discrete_data = [1, 2, 3, 4, 5]
continuous_data = [1.5, 2.3, 3.8, 4.1, 5.0]

# Data Analysis
print("Nominal Data:", nominal_data)
print("Ordinal Data:", ordinal_data)
print("Discrete Data:", discrete_data)
print("Continuous Data:", continuous_data)

# Statistical Measures
print("Mean (Discrete):", statistics.mean(discrete_data))
print("Mean (Continuous):", statistics.mean(continuous_data))
print("Mode (Discrete):", statistics.mode(discrete_data))
print("Median (Discrete):", statistics.median(discrete_data))
print("Median (Continuous):", statistics.median(continuous_data))
print("Range (Discrete):", max(discrete_data) - min(discrete_data))
print("Range (Continuous):", max(continuous_data) - min(continuous_data))
print("Skewness (Discrete):", skew(discrete_data))
print("Skewness (Continuous):", skew(continuous_data))

# Outlier Detection (using Z-score method)
def detect_outliers(data):
    z_scores = [(x - np.mean(data)) / np.std(data) for x in data]
    outliers = [x for x, z in zip(data, z_scores) if abs(z) > 3]
    return outliers

outliers_discrete = detect_outliers(discrete_data)
outliers_continuous = detect_outliers(continuous_data)

print("Outliers in Discrete Data:", outliers_discrete)
print("Outliers in Continuous Data:", outliers_continuous)

# Visualizations
plt.hist(discrete_data, bins=5)
plt.title('Histogram of Discrete Data')
plt.show()

plt.hist(continuous_data, bins=10)
plt.title('Histogram of Continuous Data')
plt.show()

plt.boxplot(discrete_data)
plt.title('Box Plot of Discrete Data')
plt.show()

plt.boxplot(continuous_data)
plt.title('Box Plot of Continuous Data')
plt.show()
```
