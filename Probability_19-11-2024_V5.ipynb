```python
# prompt: print the binomial function using random method

import pandas as pd
import numpy as np
from scipy.stats import binom

# Assuming 'x' DataFrame is already loaded as in the previous code

# Example usage (replace with your desired parameters):
n = 10000  # Number of trials
p = 0.5  # Probability of success
k = 50   # Number of successes

# Calculate binomial probability
probability = binom.pmf(k, n, p)
print(f"The probability of getting {k} successes in {n} trials with probability {p} is: {probability}")

# Generate a random variable from the binomial distribution
random_variable = binom.rvs(n, p, size=500)  # Generate 10 random variables
print(f"Random variables from the binomial distribution: {random_variable}")
```
